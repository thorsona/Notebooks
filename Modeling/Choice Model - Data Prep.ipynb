```python
### creating weight tables for choice model- demographics
query = '''
    DROP TABLE IF EXISTS thorsona.temp_choice_issue_weights;
    CREATE TABLE thorsona.temp_choice_issue_weights AS
        SELECT
            a.*,
            ROW_NUMBER() OVER () AS stratum,
            (voters / SUM(voters) OVER ()) AS pr
        FROM (
            SELECT 
                CASE WHEN a.sex = 'F' THEN 'Female'
                     WHEN a.sex = 'M' THEN 'Male'
                     ELSE 'Unknown' END AS gender,
                CASE WHEN a.age <= 34 THEN '18-34' 
                     WHEN a.age BETWEEN 35 AND 49 THEN '35-49'
                     WHEN a.age BETWEEN 50 AND 64 THEN '50-64'
                     WHEN a.age >= 65 THEN '65+'
                     ELSE 'Unknow' END AS age_group,
                CASE WHEN a.combined_ethnicity_full = 'W' THEN 'Caucasian'
                     WHEN a.combined_ethnicity_full = 'B' THEN 'African-American'
                     WHEN a.combined_ethnicity_full = 'H' THEN 'Hispanic'
                     ELSE 'Other/Unknown' END AS ethn_full,
                CASE WHEN s.coop_dem_support_score < 35 THEN 'Republican'
                     WHEN s.coop_dem_support_score BETWEEN 35 AND 65 THEN 'Unaffiliated'
                     WHEN s.coop_dem_support_score > 65 THEN 'Democrat'
                     ELSE 'None' END AS party,
                CASE WHEN t.tsmart_match_level = 'I' THEN 1 ELSE 0 END as cell_phone,
                COUNT(*) AS voters
            FROM analytics.person a
            LEFT JOIN analytics.buckets_filled x USING (personid)
            LEFT JOIN datasci.coop_dem_support_v1_1 s USING(personid)
            LEFT JOIN phones.meta_personphone p USING(personid)
            LEFT JOIN targetsmart.consumer t USING (personid)
            WHERE a.is_current_reg=True
                AND a.is_merged=False
                AND a.is_deceased=False
                AND s.coop_dem_support_score IS NOT NULL
                AND p.targetSmart_flag = 1
            GROUP BY 1,2,3,4,5
            ORDER BY 1,2,3,4,5
            ) a
        LEFT JOIN 
            (SELECT 
                gender,
                party,
                age_group,
                ethn_full
             FROM ts_national_issue_survey_combined
             WHERE choice IN ('Favor','Oppose')
             GROUP BY gender, party, age_group, ethn_full
             ) b ON
                a.gender = b.gender AND
                a.party = b.party AND
                a.age_group = b.age_group AND
                a.ethn_full = b.ethn_full
        WHERE b.ethn_full IS NOT NULL 
        ORDER BY stratum
    ;
    '''
cursor.execute(query)
```
