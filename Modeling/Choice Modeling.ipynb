```python
import argparse
import pandas as pd
import numpy as np
from dspy.modeling import helpers, model, predict, pull_data, evaluate
import time
import os
import pyodbc

conn = pyodbc.connect(dsn='VerticaThorsonA')
cursor = conn.cursor()

pd.options.display.max_rows=1000
pd.options.display.max_columns= 500

os.chdir('/home/thorsona/projects/issue_modeling_2016/choice')
params = helpers.get_params('params_dev.json')
```

```python
def pull(outcome, pull_type = 'training'):
    print 'Started pulling '+outcome+' data at '+time.strftime("%a, %d %b %Y %H:%M:%S", time.localtime())
    
    if pull_type=='training':
        pull = 'AND a.personid % 3 != 0'
    
    elif pull_type=='test':
        pull = 'AND a.personid % 3 = 0'
    
    elif pull_type=='nonparty_test':
        pull = 'AND a.personid % 3 = 0 AND reg_party_na = 1'   
    
    else:
        pull = 'AND a.personid % 3 = 0 AND reg_party_na = 0'
    
    query = '''SELECT 
                    CASE WHEN a.'''+outcome+''' = 'Favor' THEN 0 WHEN a.'''+outcome+''' = 'Oppose' THEN 1 ELSE NULL END AS outcome,
                    a.cell_flag,
                    b.*,
                    w.stratum,
                    w.pr,
                    sw.stratum as stratum_s,
                    sw.pr as pr_s
                FROM thorsona.ts_national_issue_survey_combined a 
                LEFT JOIN '''+outcome+'''_data_person_table b USING(personid)
                LEFT JOIN thorsona.temp_'''+outcome+'''_issue_weights w ON
                    a.gender=w.gender AND
                    a.age_group=w.age_group AND
                    a.ethn_full=w.ethn_full AND
                    a.party=w.party AND 
                    a.cell_flag = w.cell_phone
                LEFT JOIN thorsona.temp_issue_state_weights sw ON
                    b.state_code=sw.state_code
                WHERE a.'''+outcome+''' IN ('Favor','Oppose') '''+pull
    
    df = pd.read_sql(query, conn)
    
    data = dict()
    data['y'] = df['outcome']
    data['w'] = df[['pr','stratum','pr_s','stratum_s','ivr_flag','state_code']]
    data['x'] = df.drop(['outcome','personid', 'state_code','pr','stratum','pr_s','stratum_s','ivr_flag'], axis=1).fillna(0).select_dtypes(exclude=[object])
    
    # survey mode weights
    data['w']['ivr_weights'] = 1
    data['w'].ix[data['w'].ivr_flag==1,'ivr_weights'] = 0.5
    
    ##demographic weights
    samp_demos = pd.DataFrame(data['w'].groupby('stratum').ivr_weights.sum()/data['w'].ivr_weights.sum())
    samp_demos.columns = ['samp_demos']
    data['w'] = data['w'].join(samp_demos, on='stratum')
    data['w']['weight_demos'] = np.clip((data['w'].ivr_weights*data['w'].pr/data['w'].samp_demos), 0.5, 2)
    
    #state weights
    samp_st = pd.DataFrame(data['w'].groupby('stratum_s').ivr_weights.sum()/data['w'].ivr_weights.sum())
    samp_st.columns = ['samp_st']
    data['w'] = data['w'].join(samp_st, on='stratum_s')
    data['w']['st_weights'] = np.clip((data['w'].ivr_weights*data['w'].pr_s/data['w'].samp_st), 0.5, 2)
    
    ### combined weights
    data['w']['weights'] = sqrt(data['w']['weight_demos'] * data['w']['st_weights'])

    if pull_type=='training':
        print data['w'][['weight_demos','st_weights','weights']].describe(percentiles=[0.05,0.1,0.9,0.95])

        print '\nCell phone before weighting: {}'.format(data['w'][data['x'].cell_flag==1].weights.count() 
                        / float(data['w'].weights.count()))
        print 'Cell phone after weighting: {}'.format(data['w'][data['x'].cell_flag==1].weights.sum() 
                        / data['w'].weights.sum())

    #viewing weights by demos
    raw = pd.DataFrame(data['w'].stratum.value_counts(normalize=True))
    demo_only = raw.join(pd.DataFrame(data['w'].groupby('stratum').weight_demos.sum()/data['w'].weight_demos.sum()))
    both = demo_only.join(pd.DataFrame(data['w'].groupby('stratum').weights.sum()/data['w'].weights.sum()))
    #print np.round(100*both.join(data['w'].groupby('stratum').pr.mean()),3)

    #viewing weights by state
    raw = pd.DataFrame(data['w'].state_code.value_counts(normalize=True))
    state_only = raw.join(pd.DataFrame(data['w'].groupby('state_code').st_weights.sum()/data['w'].st_weights.sum()))
    both = state_only.join(pd.DataFrame(data['w'].groupby('state_code').weights.sum()/data['w'].weights.sum()))
    #print np.round(100*both.join(data['w'].groupby('state_code').pr_s.mean()),3)
    
    print '\nFinished pulling '+outcome+' data at '+time.strftime("%a, %d %b %Y %H:%M:%S", time.localtime())+'\n'
    return data
```

```python
#pull training data for choice model
train = pull(params['outcome_column'], 'training')
```

```python
#make sure no strata are missing, everything looks okay
train['w'].describe()
```

```python
## run grid search to pick best hyperparameters
models = {}
for model_type in params['model_types']:
    models[model_type] = model.run_gs(train, model_type, params, params['model']['params_name'].format(model_type))
    
bp = {}
for model_type in params['model_types']:
    bp[model_type] = helpers.unpickle_object(params['model']['params_name'].format(model_type), params)

## run best model incorporating the sample weights
models = {}
for model_type in params['model_types']:
    models[model_type] = model.run_best(train, model_type, params, bp[model_type], params['model']['model_name'].format(model_type),
                                        train['w'].weights)
```

```python
## pulling separate test sets for party and non-party reg states to compare how well the model performs
test_nonparty = pull(params['outcome_column'], 'nonparty_test')
test_partyreg = pull(params['outcome_column'], 'party_test')
test = pull(params['outcome_column'], 'test')
```

```python
models = {}
for model_type in params['model_types']:
    models[model_type] = helpers.unpickle_object(params['model']['model_name'].format(model_type), params)

##all scores
scores = {}
for model_type in params['model_types']:
    scores[model_type] = predict.clip(predict.get_probs(predict.predict_class_probs(test['x'], models[model_type], model_type), model_type), params)

##non-party reg states' scores
nonparty_scores = {}
for model_type in params['model_types']:
    nonparty_scores[model_type] = predict.clip(predict.get_probs(predict.predict_class_probs(test_nonparty['x'], models[model_type], model_type), model_type), params)

## party reg states' scores
party_scores = {}
for model_type in params['model_types']:
    party_scores[model_type] = predict.clip(predict.get_probs(predict.predict_class_probs(test_partyreg['x'], models[model_type], model_type), model_type), params)
```

```python
for model_type in params['model_types']:
    print 'Evaluating {} model'.format(model_type)
    preds = scores[model_type]
    evaluate.eval_model(scores=preds, outcomes=test['y'], model_name='{}'.format(model_type), params=params)
    evaluate.test_stats(scores=preds, outcomes=test['y'], model_name='{}'.format(model_type), params=params)
```

```python
##nonparty
for model_type in params['model_types']:
    print 'Evaluating {} model'.format(model_type)
    nonparty_preds = nonparty_scores[model_type]
    evaluate.eval_model(scores=nonparty_preds, outcomes=test_nonparty['y'], model_name='{}'.format(model_type), params=params)
    evaluate.test_stats(scores=nonparty_preds, outcomes=test_nonparty['y'], model_name='{}'.format(model_type), params=params)
```

```python
##party
for model_type in params['model_types']:
    print 'Evaluating {} model'.format(model_type)
    party_preds = party_scores[model_type]
    evaluate.eval_model(scores=party_preds, outcomes=test_partyreg['y'], model_name='{}'.format(model_type), params=params)
    evaluate.test_stats(scores=party_preds, outcomes=test_partyreg['y'], model_name='{}'.format(model_type), params=params)
```
